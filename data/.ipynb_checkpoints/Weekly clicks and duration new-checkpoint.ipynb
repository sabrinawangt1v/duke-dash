```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go

device=['Connector','Forlines','Vestibule1','Vestibule2','Welcomedesk']
modules=['Athletics','Campus Map','Donor Impact','DukeAroundTheWorld','Events','History','Notable Alumni']
df=pd.DataFrame(eval(open('C:/Users/Jie/Documents/dev/duke/data/week.txt', 'r').read()))
df.index=device
# find columns that satisfy: 1) not fullscreen, 2) modules that exist
bol_fullscreen=np.logical_or(df.columns.str.contains('Fullscreen'),df.columns.str.contains('FullScreen'))
bol_nomodule=df.columns.str.contains('Main')
bol1=~(np.logical_or(bol_fullscreen,bol_nomodule))
# print(df.columns)
# print(bol1)

df1=df.loc[:,bol1]

# create dataframes that contains clicks only data and duration only data
df_clicks=df1[(df1.filter(like='click').columns)]
df_duration=df1[(df1.filter(like='duration').columns)]
df_clicks.columns=modules
df_duration.columns=modules

trace1 = go.Bar(
    x=df_clicks.columns,
    y=df_clicks.iloc[0],
    name = 'Connector',
    marker=dict(color='#58508d') 
)
trace2 = go.Bar(
     x=df_clicks.columns,
    y=df_clicks.iloc[1],
    name='Welcome Desk',
    marker=dict(color='#003f5c')
)
trace3 = go.Bar(
     x=df_clicks.columns,
    y=df_clicks.iloc[2],
    name = 'Vestibule1',
    marker=dict(color='#bc5090')
)

trace4 = go.Bar(
     x=df_clicks.columns,
    y=df_clicks.iloc[3],
    name='Vestibule2',
    marker=dict(color='#ff6361') # set the marker color to coral
)


trace5 = go.Bar(
     x=df_clicks.columns,
    y=df_clicks.iloc[4],
    name = 'Forlines',
    marker=dict(color='#ffa600') 
)

data = [trace1,trace2,trace3, trace4, trace5]
layout = go.Layout(
    title='Touches of Each Module by Device',
    barmode='stack'
)
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='weekly_clicks.html')
```
